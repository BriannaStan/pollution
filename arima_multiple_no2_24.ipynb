```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import warnings
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from pandas import Series
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.preprocessing import StandardScaler
from math import sqrt
from datetime import datetime,timedelta
from dateutil.rrule import rrule, HOURLY
from statsmodels.tsa.arima.model import ARIMA
import itertools

#warnings.filterwarnings("ignore")

frames=[]
totalmape=0
no_predictions = 24
ids = [28079024, 28079038, 28079008, 28079040, 28079036, 28079018, 28079011, 28079004, 28079016, 28079039, 28079027]
for i in ids:
    dffinal=pd.read_csv("cleaned-"+str(i)+".csv", index_col='date',parse_dates=True)
    #dffinal=pd.read_csv("cleaned.csv", index_col='date',parse_dates=True)

    dffinal=dffinal.drop(columns=['NO','CH4','BEN','CO','EBE','MXY','NMHC','O_3','NOx','OXY','PM10','PXY','SO_2','TCH','TOL','PM25'])

    #print(dffinal.head())
    sc_in = MinMaxScaler(feature_range=(0, 1))
    scaled_input = sc_in.fit_transform(dffinal[['NO_2']])
    scaled_input =pd.DataFrame(scaled_input)
    X=scaled_input

    X.rename(columns={0:'NO_2'}, inplace=True)
    #X=np.log(np.log(X))

    #split data
    train_size=int(len(X) *0.7)
    test_size = int(len(X)) - train_size
    train_X = X[:train_size].dropna()
    test_X = X[train_size:].dropna()
    print("Total data size:",len(X))
    print("Train data size:",len(train_X))
    print("Test data size:",len(test_X))
    print(train_X.head())
    print(test_X.head())

    print("Index:",test_X.loc[train_size].NO_2)

    #decompose
    seas_d=sm.tsa.seasonal_decompose(X['NO_2'],model='add',period=no_predictions);
    fig=seas_d.plot()
    fig.set_figheight(4)
    plt.show()

    #corelation
    fig,ax= plt.subplots(2,1, figsize=(20,10))
    fig=sm.tsa.graphics.plot_acf(train_X, lags=50, ax=ax[0])
    fig=sm.tsa.graphics.plot_pacf(train_X, lags=50, ax=ax[1])
    plt.show()

    p = d = q = range(0,3) # p, d, and q can be either 0, 1, or 2
    pdq = list(itertools.product(p,d,q)) # gets all possible combinations of p, d, and q
    aics = [] # store aics
    # Grid Search continued
    for combination in pdq:
        try:
            mod = ARIMA(train_X, order=combination) # create all possible models
            model = mod.fit()
            aics.append([combination,model.aic])
            print('ARIMA {} x : AIC Calculated ={}'.format(combination, model.aic))
        except Exception as e:
            print(e)
            continue

    aics_df = pd.DataFrame(aics, columns=['pdq', 'aic'])
    #print(aics_df)
    best_aic = aics_df.loc[aics_df['aic'].idxmin()]
    print(best_aic)

    #fit best model
    model = ARIMA(X, order=best_aic.pdq)
    model = model.fit()
    forecast = model.forecast(no_predictions)
    print("Forecast",no_predictions,":")
    print(forecast)

    #plot residuals and density
    residuals = pd.DataFrame(model.resid)
    fig, ax = plt.subplots(1,2)
    residuals.plot(title="Residuals", ax=ax[0])
    residuals.plot(kind='kde', title='Density', ax=ax[1])
    plt.show()

    #plot prediction
    print("MAE:",model.mae)
    print("MSE:",model.mse)
    model.plot_diagnostics()
    plt.show()

    print("Forecasts:")
    print(model.forecasts)
    forecasts = pd.DataFrame(model.forecasts)
    #fig, ax = plt.subplots(1, 1)
    #forecasts.plot(title="Forecasts", ax=ax)
#    forecasts.plot(kind='kde', title='Forecast', ax=ax[1])
    plt.show();

    predictions = model.predict(start=train_size, end=train_size+no_predictions-1)
    print("Type predictions:",type(predictions))
    pred = pd.DataFrame(predictions)
    pred.index = list(pred.index)
    print("Index pred:",pred.index)
    print("Index test:",test_X.index)
    a = pred.loc[train_size]
    b = test_X.loc[train_size]
    print("A:",a)
    print("B:",b)
    print("Type pred:",type(pred))
    print("Predictions:")
    print(pred);
    tape = 0
    for i in range(train_size, train_size+no_predictions-1):
        ape = abs(pred.loc[i].predicted_mean-test_X.loc[i].NO_2)/abs(test_X.loc[i].NO_2)*100
        tape = tape + ape
    mape = tape/no_predictions;
    print("MAPE:",mape)
    print("Accuracy:",100.0-mape)
    totalmape = totalmape + mape
overallmape = totalmape/len(ids)
print("Overall MAPE:",overallmape)
print("Overall accuracy:",100.0-overallmape)
```
